```python
cv_1=cv_1[cv_1['data.customer360.data.bankaccount.aggregateaveragemonthlybalance']>0]

cv_1['accounts_json']=cv_1.apply(lambda x: json.loads(x['data.customer360.data.bankaccount.accounts']),axis=1)
cv_1['accounts_count']=cv_1.apply(lambda x: len(x['accounts_json']),axis=1)
def bank_names(x):
    banks=[]
    large_pvt=0
    large_public=0
    
    for bank in range(0, x['accounts_count']):
        if 'kotak' not in x['accounts_json'][bank]['bankName']:
            banks.append(x['accounts_json'][bank]['bankName'])
        
            
    banks=list(set(banks)) 

    if 'hdfc bank' in banks:
        large_pvt=large_pvt+1
    if 'axis bank' in banks:
        large_pvt=large_pvt+1
    if 'icici bank' in banks:
        large_pvt=large_pvt+1

    if 'state bank of india' in banks:
        large_public=large_public+1
    if 'bank of baroda' in banks:
        large_public=large_public+1
    if 'canara bank' in banks:
        large_public=large_public+1
        
    if 'punjab national bank' in banks:
        large_public=large_public+1
        
            
    return banks,large_pvt,large_public

cv_1['bank_names']=cv_1.apply(bank_names,axis=1)

```

```python

cv_1=cv_1[cv_1['data.customer360.data.bankaccount.aggregateaveragemonthlybalance']>0]
cv_1['accounts_json']=cv_1.apply(lambda x: json.loads(x['data.customer360.data.bankaccount.accounts']),axis=1)
cv_1['accounts_count']=cv_1.apply(lambda x: len(x['accounts_json']),axis=1)
cv_1['accounts_count'].value_counts()


def external_amb(x):
    month0_amb=0
    month1_amb=0
    month2_amb=0
    month3_amb=0

    month0_tot_cr=0
    month1_tot_cr=0
    month2_tot_cr=0
    month3_tot_cr=0

    month0_tot_dr=0
    month1_tot_dr=0
    month2_tot_dr=0
    month3_tot_dr=0

    for bank in range(0, x['accounts_count']):
        month0=0
        month1=0
        month2=0
        month3=0

        month0_cr=0
        month1_cr=0
        month2_cr=0
        month3_cr=0

        month0_dr=0
        month1_dr=0
        month2_dr=0
        month3_dr=0

        x['accounts_json'][bank]['bankName'].upper()
        
        if (('KOTAK' not in  x['accounts_json'][bank]['bankName'].upper())&(x['accounts_json'][bank]['bankName'].upper() not in ['SHRI VENKATESH MULTISTATE CO-OP CREDIT SOCIETY',
    'UNKNOWN',
'TRANSACTION',
'PHONEPE',
'SPAY',
'KULSOC',
'SHIVKRUPA PATPEDHI',
'PAYSWIFF',
'PAYU MONEY', 
 'KHATABOOK',
'BHIMUPI',
'SPLASH FASHION',
'AAKASH',
'MULTIPLE',
'THOMASCOOK'
'AMAZON'])):
            
            if 'month0' in x['accounts_json'][bank]['amb']:
                month0=x['accounts_json'][bank]['amb']['month0']
            
            if 'month1' in x['accounts_json'][bank]['amb']:
                month1=x['accounts_json'][bank]['amb']['month1']
            if 'month2' in x['accounts_json'][bank]['amb']:
                month2=x['accounts_json'][bank]['amb']['month2']
            if 'month3' in x['accounts_json'][bank]['amb']:
                month3=x['accounts_json'][bank]['amb']['month3']

            if 'month0' in x['accounts_json'][bank]['credit']:
                month0_cr=x['accounts_json'][bank]['credit']['month0']
            if 'month1' in x['accounts_json'][bank]['credit']:
                month1_cr=x['accounts_json'][bank]['credit']['month1']
            if 'month2' in x['accounts_json'][bank]['credit']:
                month2_cr=x['accounts_json'][bank]['credit']['month2']
            if 'month3' in x['accounts_json'][bank]['credit']:
                month3_cr=x['accounts_json'][bank]['credit']['month3']


            if 'month0' in x['accounts_json'][bank]['debit']:
                month0_dr=x['accounts_json'][bank]['debit']['month0']
            if 'month1' in x['accounts_json'][bank]['debit']:
                month1_dr=x['accounts_json'][bank]['debit']['month1']
            if 'month2' in x['accounts_json'][bank]['debit']:
                month2_dr=x['accounts_json'][bank]['debit']['month2']
            if 'month3' in x['accounts_json'][bank]['debit']:
                month3_dr=x['accounts_json'][bank]['debit']['month3']

        
        month0_amb=month0_amb+month0
        month1_amb=month1_amb+month1
        month2_amb=month2_amb+month2
        month3_amb=month3_amb+month3
        
        
        month0_tot_cr=month0_tot_cr+month0_cr
        month1_tot_cr=month1_tot_cr+month1_cr
        month2_tot_cr=month2_tot_cr+month2_cr
        month3_tot_cr=month3_tot_cr+month3_cr


        month0_tot_dr=month0_tot_dr+month0_dr
        month1_tot_dr=month1_tot_dr+month1_dr
        month2_tot_dr=month2_tot_dr+month2_dr
        month3_tot_dr=month3_tot_dr+month3_dr

    return month0_amb,month1_amb,month2_amb,month3_amb,month0_tot_cr,month1_tot_cr,month2_tot_cr,month3_tot_cr,month0_tot_dr,month1_tot_dr,month2_tot_dr,month3_tot_dr

cv_1[['ext_amb_0','ext_amb_1','ext_amb_2','ext_amb_3','ext_cr_0','ext_cr_1','ext_cr_2','ext_cr_3','ext_dr_0','ext_dr_1','ext_dr_2','ext_dr_3']]=cv_1.apply(external_amb, axis=1,result_type='expand')

```
